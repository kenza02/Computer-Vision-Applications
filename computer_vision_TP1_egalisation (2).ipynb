```python
import cv2
import matplotlib.pyplot as plt

# Lire l'image
chemin_image = '/content/dog1.jpg'  # le chemin de l'image
image = cv2.imread(chemin_image)

# Convertir l'image en niveaux de gris
image_gris = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Calculer l'histogramme
hist = cv2.calcHist([image_gris], [0], None, [256], [0,256])

# Afficher l'image en niveaux de gris
plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.title('Image originale')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Image en niveaux de gris')
plt.imshow(image_gris, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()

# Afficher l'histogramme
plt.figure()
plt.title("Histogramme de l'image en niveaux de gris")
plt.xlabel("Niveau de gris")
plt.ylabel("Nombre de pixels")
plt.bar(range(256), hist[:,0], color='black', width=1.0)
plt.xlim([0, 256])
plt.grid(True)
plt.show()
```

## -comparer le deux histogrammes

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Charger l'image en niveaux de gris
image_path = '/content/dog1.jpg'
image = cv2.imread(image_path, 0)

# Calculer l'histogramme de l'image originale
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

# Calculer la fonction de densité de probabilité (PDF)
pdf = hist / np.sum(hist)

# Calculer la fonction de répartition cumulative (CDF)
cdf = np.cumsum(pdf)

# Égalisation de l'histogramme
image_egale = cv2.equalizeHist(image)

# Calculer l'histogramme égalisé
hist_eq = cv2.calcHist([image_egale], [0], None, [256], [0, 256])

# Affichage de l'image originale avec son histogramme
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Image originale')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.plot(hist, color='black')
plt.title('Histogramme original')
plt.xlabel('Niveau de gris')
plt.ylabel('Nombre de pixels')
plt.grid(True)

# Affichage de l'image égalisée avec son histogramme
plt.subplot(2, 2, 3)
plt.imshow(image_egale, cmap='gray')
plt.title('Image égalisée')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.plot(hist_eq, color='black')
plt.title('Histogramme égalisé')
plt.xlabel('Niveau de gris')
plt.ylabel('Nombre de pixels')
plt.grid(True)

plt.tight_layout()
plt.show()
```
